```python
def compilator(path):

    # création d'un dataframe vierge
    df_comp = pd.DataFrame()
    
    # pour chaque fichier du dossier path
    for file in os.listdir(path):
        # si fichier csv
        if file.endswith('.csv'):
            # création du dataframe à partir du fichier csv
            df = pd.read_csv(path + file, header=20)
            
            # récupération du numéro de la newsletter
            digest_num = int(file[25:27].strip("_"))
            # ajout de ce num dans la colonne digest
            df['digest'] = digest_num
            
            # compilation du dataframe
            df_comp = df_comp.append(df, ignore_index=True)
    
    # renommage des colonnes
    df_comp.columns = ["url", "total_clicks", "unique_clicks", "digest_num"]
    # triage du dataframe
    df_comp.sort_values(by=["digest_num", "total_clicks", "unique_clicks", "url"],
                        ascending=[True, False, False, True],
                        inplace=True)
    # recréation de l'index
    df_comp.reset_index(inplace=True, drop=True)
    
    return df_comp
    
```

```python
reports_data = compilator(reports_path)
reports_data
```
